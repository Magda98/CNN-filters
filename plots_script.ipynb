```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'svg'
```

```python
def plots(methods, title):
    for m in methods:
        temp = np.empty((3, 200), np.float64)
        for i in range(3):
            t = np.array([])
            with open('./data_plots/'+m+str(i+10)+'.csv') as f:
                for l in f:
                    t = np.append(t, float(l))
            temp[i] = t
        label = m[5:] if "cifar" in m else m
        mean = np.mean(temp, axis=0)
        plt.plot(range(200), mean, label=label)

    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend(loc='best')
    # plt.title(label=title, pad=20.0)
    plt.show()
```

```python
methods_intel = ["xavier_uniform", "xavier_uniform_M_1", "xavier_uniform_M_2", "xavier_uniform_M_10", "xavier_uniform_M_14", "xavier_uniform_M_20"]
plots(methods=methods_intel, title="Wykres błędu na wyjściu sieci w trakcie procesu uczenia dla zbioru Intel")
```

```python
methods_cifar = ["cifarxavier_uniform", "cifarxavier_uniform_M_1", "cifarxavier_uniform_M_2", "cifarxavier_uniform_M_10", "cifarxavier_uniform_M_14", "cifarxavier_uniform_M_20"]
plots(methods=methods_cifar, title="Wykres błędu na wyjściu sieci w trakcie procesu uczenia dla zbioru Cifar-10")
```
