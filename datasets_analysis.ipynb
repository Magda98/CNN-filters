```python
from torchvision import datasets
import pandas as pd
from torch.utils.data import DataLoader
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'svg'

train_data = datasets.ImageFolder('./datasets/cifar10/train')
test_data = datasets.ImageFolder('./datasets/cifar10/test')

(idx, class_count) = torch.unique(torch.tensor(train_data.targets), return_counts=True)
idx = idx.numpy()
class_count = class_count.numpy()

colors = sns.color_palette('PuBu', 10)
plt.pie(class_count, labels=train_data.classes, colors=colors, autopct='%.0f%%', explode=[0.02]*len(class_count))
plt.show()
```

```python
dataset = [len(train_data.targets), len(test_data.targets)]
colors = sns.color_palette('PuBu', 2)

plt.pie(dataset, labels=["Dane treningowe", "Dane testowe"], colors=colors, autopct='%.0f%%', explode=[0.02]*2)
plt.show()
```
